````python
#1. Define a function that given a pandas DataFrame as input creates a **seaborn countplot** of each categorical column. 
#Make sure to sort the bars by frequency ie: the most frequent values should be placed first. Hint: use .value_counts(). 
#In addition, if the amount of unique values of a categorical column (cardinality) is six or more, the corresponding countplot should have the bars placed on the y-axis instead of the x-axis.
````

````python
def plot_categorical_columns(df, columns=None, max_categories=16, figsize=(10, 6)):
    # If no specific columns are provided, use all categorical columns
    if columns is None:
        categorical_columns = df.select_dtypes(include=['object', 'category']).columns
    else:
        categorical_columns = [col for col in columns if col in df.columns and df[col].dtype in ['object', 'category']]

    if len(categorical_columns) == 0:
        print("No categorical columns to plot.")
        return

    for column in categorical_columns:
        value_counts = df[column].value_counts()

        # Limit the number of categories to display
        if len(value_counts) > max_categories:
            value_counts = value_counts.iloc[:max_categories]

        plt.figure(figsize=figsize)

        if len(value_counts) >= 6:
            sns.countplot(y=column, data=df, order=value_counts.index)
        else:
            sns.countplot(x=column, data=df, order=value_counts.index)

        plt.title(f'Countplot of {column}')
        plt.tight_layout()
        plt.show()
````

````python
plot_categorical_columns(df)
````

````python
#13. Can you do a monthly analysis given the output of the previous series? Why?
````

````python
#Limited Data Coverage: The series only contains data for January and February. Without data for the other months (March to December), any analysis would be incomplete and not representative of the entire year.

#Lack of Comparability: Monthly analysis often involves comparing trends, patterns, and changes over different months. With only two months of data, such comparisons are highly limited.

#Seasonal and Monthly Variations: Many datasets exhibit seasonal trends or variations that can only be understood by analyzing data across all months. With only a subset of the year covered, it's impossible to capture these seasonal effects or make meaningful inferences about them.

#Statistical Significance: The reliability of any statistical analysis is dependent on having a sufficiently large and representative dataset. The absence of data for most of the year limits the ability to draw statistically significant conclusions.

#In summary, while you can analyze and draw conclusions from the available data for January and February, these insights will be limited to just these two months. For a full monthly analysis that covers seasonal trends, comparisons between months, and statistically significant conclusions, data for all months of the year would be necessary.
````

````python
def monthly_analysis(df):
    monthly_data = df.groupby(df['effective_to_date'].dt.month_name())['total_claim_amount'].sum().round(2)
    return monthly_data
````

````python
#14. Define a function to remove the outliers of a numerical continuous column depending if a value is bigger or smaller than a given amount of standard deviations of the mean (thr=3).
````

````python
def remove_outliers(new_df, column, thr=3):
    mean = new_df[column].mean()
    std_dev = new_df[column].std()
    lower_limit = mean - (thr * std_dev)
    upper_limit = mean + (thr * std_dev)

    filtered_df = new_df[(new_df[column] >= lower_limit) & (new_df[column] <= upper_limit)]

    return filtered_df
````

````python
continuous_cleaned_df = new_df.copy()
for column in ['customer_lifetime_value', 'total_claim_amount', 'income', 'monthly_premium_auto']:
    continuous_cleaned_df = remove_outliers(continuous_cleaned_df, column)
````
